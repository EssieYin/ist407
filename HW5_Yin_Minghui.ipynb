# Question 2
Imagine you are a running bike sharing business, then you want to predict the rental count based on the environmental & seasonal settings by using regression tree algorithm. You need to predict both hourly and daily rental counts (total counts). The hourly data and daily data are in two files: hour.csv and day.csv.  
    
Is there any difference in terms of predictability? Hint: ignore the instant and dteday columns. And our dependent variable is the total bike rental count, which means you can not include casual and registered counts into the feature set.   

source: https://archive.ics.uci.edu/ml/datasets/Bike+Sharing+Dataset

One more column in hour dataset.    (which is hr)   
- instant: record index
- dteday : date
- season : season (1:winter, 2:spring, 3:summer, 4:fall)
- yr : year (0: 2011, 1:2012)
- mnth : month ( 1 to 12)
- hr : hour (0 to 23)
- holiday : weather day is holiday or not (extracted from [Web Link])
- weekday : day of the week
- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
+ weathersit :
    - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp : Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)
- atemp: Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)
- hum: Normalized humidity. The values are divided to 100 (max)
- windspeed: Normalized wind speed. The values are divided to 67 (max)
- casual: count of casual users
- registered: count of registered users
- cnt: count of total rental bikes including both casual and registered

**R-Squared is a statistical measure of fit that indicates how much variation of a dependent variable is explained by the independent variable(s) in a regression model.**

```python
# Tuning Decision Tree Model

RSquared_train = []
RSquared_test = []
min_samples_leaf = []
max_depth = []

for i in range(5,21):
    for j in range(1,50):
        tree_reg = DecisionTreeRegressor(min_samples_leaf= i, random_state= 2022, max_depth = j)
        tree_reg.fit(X_train, y_train)
        R2_train = metrics.r2_score(y_train, tree_reg.predict(X_train))
        R2_test = metrics.r2_score(y_test, tree_reg.predict(X_test))
        RSquared_train.append(R2_train)
        RSquared_test.append(R2_test)
        min_samples_leaf.append(i)
        max_depth.append(j)

R_Squared = pd.DataFrame()
R_Squared["RSquared_train"] = RSquared_train
R_Squared["RSquared_test"] = RSquared_test
R_Squared["min_sample_leaf"] = min_samples_leaf
R_Squared["max_depth"] = max_depth

R_Squared.head()
```
